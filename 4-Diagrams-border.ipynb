# Diagrams

```python
#import the pandas library & alias as pd
import pandas as pd

# load the data into a DataFrame
grades = pd.read_csv("grades.csv")

# print the DataFrame
print(grades)
```

<div style="float: left;width: 45%; border-right: 1px solid black">
    
We can use pandas in combination with Matplotlib to visualize<br> DataFrames.
<br><br>The <code>plot()</code> function creates diagrams. 
</div>

<div style="float: right;width: 50%">

We kunnen Pandas gebruiken in combinatie met Matplotlib om DataFrames te visualiseren. 
<br><br>Met de functie <code>plot()</code> worden diagrammen gemaakt. 
</div>

```python
# import matplotlib and alias as plt
import matplotlib.pyplot as plt

grades.plot()

plt.show()
```

<div style="float: left;width: 45%; border-right: 1px solid black; padding-right: 2%">
    
We can use the <code>kind</code> argument to choose the type of plot.
<br><br>
<h3> scatter plot </h3>
<br>Let's create a scatter plot: 
</div>

<div style="float: right;width: 50%">

We kunnen het argument <code>kind</code> gebruiken om het type van de plot te kiezen. 
<h3> scatter plot </h3>
<br>We maken een spreidingsplot: 
</div>

```python
# scatter plot
grades.plot(kind="scatter", x="lab", y="theory")

plt.show()
```

<div style="float: left;width: 45%; border-right: 1px solid black; padding-right: 2%">
    
<h3> histogram </h3>
<br>Let's create a histogram:  
</div>

<div style="float: right;width: 50%">

<h3> histogram </h3>
<br>We maken een histogram:  
</div>

```python
# histogram

grades["final"].plot(kind="hist")

plt.show()
```

```python
# histogram with multiple columns

grades[["lab","theory"]].plot(kind="hist")

plt.show()
```

<div style="float: left;width: 45%; border-right: 1px solid black; padding-right: 2%">
    
<h3> bar chart </h3>
<br>Let's create a bar chart: 
</div>

<div style="float: right;width: 50%">

<h3> bar chart </h3>
<br>We maken een staafdiagram:  
</div>

```python
# bar chart with multiple columns

grades[["lab", "theory"]].plot.bar()

plt.show()
```

```python
# stacked bar chart

grades[["lab", "theory"]].plot.bar(stacked=True)
```

<div style="float: left;width: 45%; border-right: 1px solid black; padding-right: 2%">
    
<h3> box </h3>
<br>Let's create a box diagram: 
</div>

<div style="float: right;width: 50%">

<h3> box </h3>
<br>We maken een boxdiagram   
</div>

```python
# box
grades.plot.box()

plt.show()
```

<div style="float: left;width: 45%; border-right: 1px solid black; padding-right: 2%">
    
<h4> replace empty values </h4>
<br>We can deal with empty cells by replacing empty values. 
<br><br>The <code>fillna()</code> method allows us to replace empty cells with a value. 
<br><br>Similarly to <code>dropna()</code>,  <code>fillna()</code> returns a new DataFrame, and will NOT change the original. 
<br><br>If we want to modify the original DataFrame, we use the <code>inplace = True</code> argument. 
</div>

<div style="float: right;width: 50%">

<h4> Lege waarden vervangen </h4>
<br>We kunnen met lege cellen omgaan door de lege waarden te vervangen.  
<br><br>Met de <code>fillna()</code> functie kunnen we lege cellen met een waarde vervangen. 
<br><br>Net als <code>dropna()</code>, geeft <code>fillna()</code>een nieuw DataFrame terug, en zal het origineel NIET vervangen. 
<br><br>Als we het originele DataFrame willen veranderen, gebruiken we het argument <code>inplace = True</code>.   
</div>
